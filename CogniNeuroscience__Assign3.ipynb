# Q1. Step current input


a) Plot the membrane voltage when the following step current input is given to the neuron with 
- Resting potential: -70 mV, 
- Membrane resistance: 10^7ohm, 
- Membrane time-scale: 8 ms (simulation time = 200 ms)
- T_start = 25 ms
- T_end = 200 ms
- Current Amplitude = 2 nA

```
firingThreshold = 200 # in mV
step_current = input_factory.get_step_current(
    t_start=25, t_end=200, unit_time=b2.ms,
    amplitude= 2 * b2.nA)

(state_monitor, spike_monitor) = LIF.simulate_LIF_neuron(input_current=step_current, 
                                                         simulation_time = 200 * b2.ms,
                                                         v_rest= -70 * b2.mV,
                                                         firing_threshold = firingThreshold * b2.mV,
                                                         membrane_resistance = 10e7 * b2.ohm,
                                                         membrane_time_scale = 8*b2.ms)
plot_tools.plot_voltage_and_current_traces(
state_monitor, step_current, title="Membrane Voltage vs Step Current", firing_threshold=firingThreshold*b2.mV)
plt.show()
```

d) Plot the response of the neuron when a step current of different amplitudes[1nA, 2nA, 4 nA, 8 nA, 16 nA] is given.Comment on the response of the neuron w.r.t. the Passive Membrane equation.

e) A step input current is given to a Passive Membrane neuron which is at a higher potential than its resting membrane potential. What is the difference in response of this neuron as compared to a neuron which is given a similar step input current at its resting potential? Assume all the other neuron parameters and variables are same for both the neurons.

**Answer:**

When a step current is given to neuron at resting potential, it reaches a steady state potential.

When a step current input is given to membrane at a higher potential than its resting potential, it would reach a higher steady state potential than it would have reached when it was at its resting potential. In short, the potential adds up and reaches a higher steady state potential.

```
firingThreshold = 500 # in mV

sim_time = 200

curr_arr = np.zeros((sim_time, 1)) * b2.nA
curr_arr[49 : 70, 0] = 2 * b2.nA
curr_arr[70 : , 0] = 4 * b2.nA

step_current = b2.TimedArray(curr_arr, dt=1.0 * b2.ms)

(state_monitor, spike_monitor) = LIF.simulate_LIF_neuron(input_current=step_current, 
                                                         simulation_time = 200 * b2.ms,
                                                         v_rest= -70 * b2.mV,
                                                         firing_threshold = firingThreshold * b2.mV,
                                                         membrane_resistance = 10e7 * b2.ohm,
                                                         membrane_time_scale = 8*b2.ms)
plot_tools.plot_voltage_and_current_traces(
      state_monitor, step_current, title="Response when step current is given to a neuron at a higher potential ", firing_threshold=firingThreshold*b2.mV)
plt.show()
```

# Q3. Multiple pulse current inputs

- Pulse1: T_start = 80 ms, T_end = 100 ms, Current amplitude = 4 nA
